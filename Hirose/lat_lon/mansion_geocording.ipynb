```python
import googlemaps
import pandas as pd

df = pd.read_csv('mansion.csv')

gm = googlemaps.Client(key='xxxxxxxxxx')
for i, r in df.iterrows():
    res = gm.geocode(r['adress'])
    df.loc[i,'latitude'] = res[0]['geometry']['location']['lat']
    df.loc[i,'longitude'] = res[0]['geometry']['location']['lng']
    print(i)
    print(res[0]['geometry']['location'])
```

```python
df.to_csv('mansion.csv', index=None)
```
