# 区分の列にあるすべての区分をリスト化する

```python
type_list= list(df['区分'].unique())
```

# マルチセレクトの作成

```python
selections= st.multiselect("ご希望の保育園区分をお選びください", type_list, default='認可保育園')
```


# セレクトされた区分でデータフレームの中身をフィルタリングする

```python
df2 = df[(df['区分'].isin(selections))]
```

# 地図の作成
プロットするためのベースの地図を作成するコードにあたる。表示させる地図の中心地を指定。OpenstreetMapは無料で使える。zoom_startはズーム倍率を指定できる。
詳しくは、 参考URL：https://maps.gsi.go.jp/development/ichiran.html#pale

```python
map = folium.Map(
    # 地図の中心位置の指定(今回は世田谷区玉川地区の玉川総合支所を指定)
    location=[35.608730497916845, 139.64832587827124],
    tiles="OpenStreetMap",
    zoom_start=13
)
```

# 保育園の区分（マルチセレクト）とプロットを連動させる
folium.Circleは丸のプロット。  
df2 = df[(df['区分'].isin(selections))]の区分毎のデータをfor構文で回し、.add_to(map)で先ほど作ったベースの地図に追加しプロットさせる。  
参考URL：https://discuss.streamlit.io/t/how-do-i-add-multi-select-to-my-folium-map/22683　　

ここでカラーやアイコン（ここでは一括で●だが）をカスタマイズできる。  
location・・・・緯度経度  
popup・・・ポップアップ  
radius・・・丸の角度  
fill・・・丸の中を埋めるかどうか  
color・・・カラー  


```python
for i, row in df2.iterrows():
    iframe = folium.IFrame('・区分: ' + row.loc['区分'] + '<br>' + '・名称:' + row.loc['名称']+ '<br>' + '・定員:' + row.loc['定員']+'<br>' + '・1歳定員:' + str(row.loc['1歳定員']) )
    popup = folium.Popup(iframe, min_width=300, max_width=300)
    folium.Circle(
        location= [row["latitude"], row["longitude"]],
        tooltip= row["名称"],
        popup= popup,
        radius= 10,
        fill= True,
        color= 'red'
    ).add_to(map)
```

```python
st_folium(map, width=700, height=700)
```
